```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
poly_x=poly.fit_transform(x)
```

```python
reg2=LinearRegression()
```

```python
reg2.fit(poly_x, y)
```

```python
reg2.predict(poly.fit_transform([[6.5]]))
```
